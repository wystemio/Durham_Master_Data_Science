```python
# CODE GOES HERE  
# Generate lists contains random integer data 

def dot_pro(lsta, lstb): 
    if len(lsta) != len(lstb):  # if we use Q1 function, the length will be always equal. 
        return 'Please enter two lists with same length'
    elif len(lsta) == 0:
        return None
    else:
        dotpro = 0
        for i in range(len(lsta)):
            dotpro = dotpro + lsta[i] * lstb[i]
        return dotpro

lstc, lstd = lst_get(0, 20, 10)
answer_Q2 = dot_pro(lstc, lstd)

print(f'lstc = {lstc}',f'lstd = {lstd}', sep = '\n')  # for double-checking
print(answer_Q2)
```

9. Write a function that takes an ‘n x n’ matrix as an argument and applies the built-in NumPy determinant function to that matrix. [5%]
